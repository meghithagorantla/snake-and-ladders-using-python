```python
import random

numbers=100

   
def check_ladder(pts):
    if(pts==2):
        print('Ladder')
        return 23
    elif(pts==6):
        print('Ladder')
        return 45
    elif(pts==20):
        print('Ladder')
        return 59
    elif(pts==52):
        print('Ladder')
        return 72
    elif(pts==57):
        print('Ladder')
        return 96
    elif(pts==71):
        print('Ladder')
        return 92
    else:
        return pts
   
 
def check_snake(pts1):
    if(pts1==43):
        print('snake')
        return 17
    elif(pts1==50):
        print('snake')
        return 5
    elif(pts1==56):
        print('snake')
        return 8
    elif(pts1==73):
        print('snake')
        return 15
    elif(pts1==84):
        print('snake')
        return 58
    elif(pts1==87):
        print('snake')
        return 49
    elif(pts1==98):
        print('snake')
        return 40
    else:
        return pts1
   

def reached_numbers(pts):
    if pts==numbers:
        return True
    else:
        return False
       
   
def play():
     p1_name=input('player1, please enter ur name:')    
     p2_name=input('player2, please enter ur name:') 
   
         
     
     p1=0
     p2=0
     turn=0
    
     while True:
            
            
            
            if turn==0:
                 
                print(p1_name,'your turn')
             
             
                 
             
                dice=random.randint(1,6)
                print('dice showed',dice)
                p1=p1+dice
                p1=check_ladder(p1)
                p1=check_snake(p1)
            if p1>numbers:
                
                p1=numbers
                print(p1_name,"ur score",p1)
            if reached_numbers(p1):
                
                print(p1_name,'Won')
                break
            else:
                print(p2_name,'your turn')
            
                         
                dice=random.randint(1,6)
                print('dice showed',dice)
                p2=p2+dice
                p2=check_ladder(p2)
                p2=check_snake(p2)
            if p2>numbers:
                 p2=numbers
            print(p2_name,"ur score",p2)
            if reached_numbers(p2):
                print(p2_name,'Won')
                break    
            turn=turn+1    
             
          

play()
```
